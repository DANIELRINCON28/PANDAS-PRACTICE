```python
import numpy as np
datos.replace('?', np.nan, inplace=True) #reemplazar '?' por NaN
```

```python
datos.info()
```

```python
datos['normalized-losses'].value_counts()
```

```python
datos['normalized-losses'] = datos['normalized-losses'].astype('float64')
```

```python
media = datos['normalized-losses'].mean()
datos['normalized-losses'].replace(np.nan, media, inplace=True)
```

```python
datos.info()
```

```python
datos['num-of-doors'].value_counts()
```

```python
moda = datos['num-of-doors'].mode()[0]
moda
```

```python
datos['num-of-doors'].fillna(moda, inplace=True)
```

```python
datos['horsepower'].value_counts()
```

```python
datos['bore'] = datos['bore'].astype('float64')
media = datos['bore'].mean()
datos['bore'].replace(np.nan, media, inplace=True)
```

```python
datos.info()
```

```python
datos['stroke'] = datos['stroke'].astype('float64')
media = datos['stroke'].mean()
datos['stroke'].replace(np.nan, media, inplace=True)
```

```python
datos.info()
```

```python
datos['horsepower'] = datos['horsepower'].astype('float64')
media = datos['horsepower'].mean()
datos['horsepower'].replace(np.nan, media, inplace=True)
```

```python
datos.info()
```

```python
datos['peak-rpm'] = datos['peak-rpm'].astype('float64')
media = datos['peak-rpm'].mean()
datos['peak-rpm'].replace(np.nan, media, inplace=True)
```

```python
datos['price'] = datos['price'].astype('float64')
media = datos['price'].mean()
datos['price'].replace(np.nan, media, inplace=True)
```

```python
datos.info()
```

```python
datos['city-kpg'] = datos['city-mpg'] * 1.60934
datos.head()
```
